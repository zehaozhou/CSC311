```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 15, 16)
bias = (x**2)/(x + 1)**2
variance = 9/(10 * (x + 1)**2)
error = (10 * x**2 + 9)/(10 * (x + 1)**2)
plt.plot(x, bias, 'go--', label='bias')
plt.plot(x, variance, 'bo--', label='variance')
plt.plot(x, error, 'ro--', label='error')
plt.legend()
plt.xlabel("$\lambda$")
plt.show()
```
